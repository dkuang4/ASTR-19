```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create a random binomial distribution
#np.random.binomial(n,m,size) returns array of specified size with
#number of successes within n trials and m success rate
x = np.random.binomial(100, 0.25, size=1000)

#create the histogram
plt.hist(x, bins=100, alpha=0.5, edgecolor='black')
plt.xlabel('x')
plt.ylabel('N per bin')

#save the figure
plt.savefig('code-journal-2-4.pdf', bbox_inches='tight', facecolor='white')
```
