```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create an array of randomly distributed numbers between 0 and 1
#uniform distribution, not normal distribution (Gaussian)
x = np.random.uniform(size=1000)

#plot the histogram
plt.hist(x, bins=100, alpha=0.5, edgecolor='black')

#labels
plt.xlabel('x')
plt.ylabel('N per bin')

#save figure
plt.savefig('code-journal-2-3.pdf', bbox_inches='tight', facecolor='white')
```
