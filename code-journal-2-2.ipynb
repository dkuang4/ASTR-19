```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#define array x as an array 0 to 1 in 100 steps
x = np.linspace(0,1,100)
```

```python
#define functions that return sin(x) and cos(x)
#n is used as a parameter instead of x to avoid confusion, but x can still be argument later
def sin(n):
    return np.sin(n)

def cos(n):
    return np.cos(n)
```

```python
#making a multipanel figure
f, axarr = plt.subplots(1,2)

#first figure
axarr[0].plot(x, sin(x))
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title(r'$sin(x)$')

axarr[1].plot(x, cos(x))
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$cos(x)$')

f.subplots_adjust(wspace=0.4)

plt.savefig('code-journal-2-2.pdf', bblox
```
